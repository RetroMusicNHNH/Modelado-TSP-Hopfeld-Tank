# 01 - Exploración de las instancias

Visualizamos las dos instancias del TSP, sus matrices de distancias y las rutas óptimas calculadas por fuerza bruta (valores de referencia del informe).

```python
import sys
from pathlib import Path

ROOT = Path.cwd()
if ROOT.name == 'notebooks':
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT))

import numpy as np
import matplotlib.pyplot as plt

from src.instances import FOUR_CITIES, FIVE_CITIES, distance_matrix_for
from src.brute_force import brute_force_optimal
from src.analysis import route_as_cycle_string
from src.visualization import plot_route
```

## Matrices de distancias

```python
for instance in (FOUR_CITIES, FIVE_CITIES):
    d = distance_matrix_for(instance)
    print(f'== {instance.name} ==')
    print(np.round(d, 2))
    print()
```

## Rutas óptimas por fuerza bruta

```python
fig, axes = plt.subplots(1, 2, figsize=(11, 5))
for ax, instance in zip(axes, (FOUR_CITIES, FIVE_CITIES)):
    d = distance_matrix_for(instance)
    opt = brute_force_optimal(d)
    cycle = route_as_cycle_string(opt.route, instance)
    plot_route(instance.coords, opt.route, labels=instance.labels,
               title=f'{instance.name}: {cycle} (L={opt.length:.2f})', ax=ax)
plt.tight_layout()
plt.show()
```
